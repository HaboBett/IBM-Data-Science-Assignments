```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
%matplotlib inline
mpl.style.use(['ggplot'])
```

```python
coffee_df = pd.read_csv('Coffee_domestic_consumption.csv')
coffee_df.head()
```

```python
coffee_df.info()
```

```python
coffee_df.columns.dtype
```

```python
coffee_df.isnull().sum()
```

```python
coffee_df.tail()
```

```python
coffee_df.describe()
```

```python
coffee_df['Coffee type'].value_counts()
```

```python
coffee_df['Coffee type'] = np.where(coffee_df['Coffee type']=='Robusta/Arabica', 'Arabica/Robusta', coffee_df['Coffee type'])
```

```python
coffee_df['Coffee type']
```

```python
coffee_df['Coffee type'].value_counts()
```

```python
coffee_df.columns
```

```python
coffee_df.set_index('Country', inplace=True)
```

```python
coffee_df.head()
```

```python
ken_df = coffee_df.loc['Kenya', '1990/91':'2019/20']
ken_df
```

```python
ken_df.plot(kind='line', figsize=(18,9))
plt.title('Coffee Consumption in Kenya from 1990/91 - 2019/20')
plt.xlabel('Year of consumption')
plt.ylabel('Total consumption')
plt.annotate('', xy=(26,4.2e6), xytext=(22, 3.2e6), xycoords='data', arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))
plt.annotate('Sharp increase in consumption', xy=(23.5, 3.5e6),rotation=71, va='baseline', ha='center')
```

```python
east_africdf = coffee_df.loc[['Kenya', 'Uganda', 'Tanzania'], '1990/91':'2019/20']
east_africdf
```

```python
east_africdf = east_africdf.T
```

```python
east_africdf.plot(kind='area', stacked=False, alpha=0.45, figsize=(14,7))
plt.xlabel('Consumption span')
plt.ylabel('Amount of consumption')
plt.title('Comparison of coffee consumption in South America')
```

```python
coffee_df.columns
```

```python
coffee_df.sort_values(by='Total_domestic_consumption', ascending=False, inplace=True)
coffee_df
```

```python
colombia_df = coffee_df.loc[['Colombia', 'Venezuela'], '1990/91':'2019/20']
colombia_df = colombia_df.T
colombia_df.index
```

```python
#counts, bin_edges = np.histogram(colombia_df)
colombia_df.plot(kind='area',  figsize=(17,8), alpha=0.45)

plt.xlabel('Years of Consumption')
plt.ylabel('Amount of Consumption')
plt.xlim(0,19)
plt.title('Comparison in trend of consumption between Venezuela and Colombia')
```

```python
coffee_df.columns = coffee_df.columns.map(lambda x: x.split('/')[0])
```

```python
coffee_df.columns
```

```python
years = list(map(lambda x: str(x), range(1990,2020)))
venezuela_df = coffee_df.loc['Venezuela', years]
venezuela_df
```

```python
count, bin_edges = np.histogram(venezuela_df, bins=15)
venezuela_df.plot(kind='bar', figsize=(17,8))
plt.xlabel('Consumption year')
plt.ylabel('Quantity of consumption')
plt.title('Coffee consumption in Venezuela from 1990 - 2019')
```

```python
type_distribution = coffee_df.loc[:, ['Coffee type', 'Total_domestic_consumption']]
```

```python
type_group = type_distribution.groupby(['Coffee type']).mean()
type_group.index.name = None
```

```python
type_group.plot(kind='bar', figsize=(10,5))
plt.ylabel('Total consumption')
plt.title('Coffee Consumption by type')
```

```python
df_top15 = coffee_df.head(15)
df_top15
```

```python
df_top15['Total_domestic_consumption']
```

```python
df_top15['Total_domestic_consumption'].plot(kind='barh', figsize=(17,8))
plt.title('Total Coffee Consumption in the top 15 countries')
plt.xlabel('Coffee consumption')
```

```python
df_2010 = coffee_df.loc[:, '2010']
df_2010
```

```python
counts, bin_edges = np.histogram(df_2010)
df_2010.plot(kind='hist', xticks=bin_edges, figsize=(17,8))
plt.xlabel('Coffee consumption')
plt.ylabel('No. of countries')
plt.title('Histogram of Coffee Consumption by Country')
```

```python
tot_df = coffee_df.loc[['Brazil', 'Venezuela', 'Indonesia', 'Mexico'], years]
```

```python
tot_df = tot_df.T
tot_df
```

```python
counts, bin_edges = np.histogram(tot_df)
tot_df.plot(kind='hist', xticks=bin_edges, figsize=(17,5), alpha=0.45)
plt.xlabel('Consumption bins')
plt.ylabel('No. of years')
plt.title('Distribution of Coffee Consumption by year')
plt.annotate('', xy=(1.332e9, 9), xytext=(0.432e9, 9), arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='black', lw=2))
plt.annotate('Brazil Domination in Coffee Consumption', xy=(1.14e9, 10), xytext=(0.561e9, 10), va='bottom', ha='left')
```

```python
east_africdf = coffee_df.loc[['Kenya', 'Uganda', 'Tanzania'], years]
east_africdf
```

```python
east_africdf = east_africdf.T
east_africdf
```

```python
east_africdf.plot(kind='bar', figsize=(17,8))
plt.xlabel('Year')
plt.ylabel('Total Consumption')
plt.title('Comparison of Coffee Consumption between Kenya, Uganda and Tanzania.')
plt.annotate('', xy=(27, 1.5e7), xytext=(18, 1.2e7), xycoords='data', arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='black', lw=2))
plt.annotate('Steady increase in Ugandan consumption',xy=(27,1.5e7), xytext=(19,1.25e7), rotation=17, va='bottom', ha='left')
```

```python
counts, bin_edges = np.histogram(east_africdf)
east_africdf.plot(kind='hist', xticks=bin_edges, alpha=0.45, figsize=(17,8))
plt.xlabel('Consumption bins')
plt.ylabel('No. of years')
plt.title('Comparison of Consumption Distribution')
```
